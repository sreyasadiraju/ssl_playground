```python
criterion_sl = torch.nn.CrossEntropyLoss()
```

```python
optim_sl = torch.optim.Adam(model_sl.parameters(), lr=1e-3)
optim_ssl = torch.optim.Adam(model_ssl.parameters(), lr=1e-3)
```

```python
epochs = 2
log_freq = 20

for epoch in range(epochs):
    
    running_loss = 0.0
    running_time = time.time()
    
    for step, (imgs, tgts) in enumerate(mnist64.train_data_loader):
                
        tfmed_imgs, _ = MNIST64.random_transform_images_uniform(imgs)
        outputs = model_sl(tfmed_imgs)
        
        loss = criterion_sl(outputs, tgts)
        loss.backward()
        optim_sl.step()
        optim_sl.zero_grad()
        
        running_loss += loss
        
        if (step+1) % log_freq == 0:
            print(f"[Epoch {epoch+1}, Step {step+1}] Loss: {running_loss/log_freq:.3f}, Time per step: {1000*(time.time() - running_time)/log_freq:.0f} ms")
            running_loss = 0.0
            running_time = time.time()
        
        
```

```python
def off_diagonal(x):
    # return a flattened view of the off-diagonal elements of a square matrix
    n, m = x.shape
    assert n == m
    return x.flatten()[:-1].view(n - 1, n + 1)[:, 1:].flatten()
```

```python
lambd = 0.0051

def barlow_twins_loss(outputs1, outputs2):
    
    c = outputs1.T @ outputs2

    on_diag = torch.diagonal(c).add_(-1).pow_(2).sum()
    off_diag = off_diagonal(c).pow_(2).sum()
    return on_diag + lambd * off_diag
```

```python
epochs = 2
log_freq = 20

for epoch in range(epochs):
    
    running_loss = 0.0
    running_time = time.time()
    
    for step, (imgs, _) in enumerate(mnist64.train_data_loader):
                
        tfmed_imgs1, tfmed_imgs2, _ = MNIST64.random_transform_pairs_uniform(imgs)
        outputs1 = model_ssl(tfmed_imgs1)
        outputs2 = model_ssl(tfmed_imgs2)
        
        loss = barlow_twins_loss(outputs1, outputs2)
        loss.backward()
        optim_ssl.step()
        optim_ssl.zero_grad()
        
        running_loss += loss
        
        if (step+1) % log_freq == 0:
            print(f"[Epoch {epoch+1}, Step {step+1}] Loss: {running_loss/log_freq:.3f}, Time per step: {1000*(time.time() - running_time)/log_freq:.0f} ms")
            running_loss = 0.0
            running_time = time.time()
```
